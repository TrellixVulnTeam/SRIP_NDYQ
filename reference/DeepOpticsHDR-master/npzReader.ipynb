```python
def PSFnpzReader(filePath):
    data = np.load(filePath)
    lst = data.files
    return data[lst[0]]

def PSFviewer(filePath):
    PSF = PSFnpzReader(filePath)
    PSF = PSF[:,:,0,:]
    PSF = (PSF - PSF.min()) / (PSF.max() - PSF.min()) * 255
    plt.imshow(PSF, cmap='gray')
```
